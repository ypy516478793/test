# Literature review

Review papers related to lung cancer screening using deep learning methods

### What is [Incidental Lung Nodule problem](https://radiopaedia.org/articles/incidental-lung-nodules-1)?
The risk of developing cancer in very small nodules(<5 mm) is very low. However, clear-cut recommendations are sill not in place with high variation in practice amongst reporting radiologist. As a result, it is common practice to note these nodules in the report and move on, leaving the ball in the clinician's court to decide whether to follow them up and at what interval.

### Goal
> **Develop protocols and tools for using CT (volumetry) of lung nodules as an quantitiative imaging biomarkers of cancer risk.**

#### [End-to-end lung cancer screening with three-dimensional deep learning on low-dose chest computed tomography](natmed19letter.pdf)

**Authors:** Diego Ardila et al.   
**Institutes:** Google AI; Stanford Health Care; Northwestern Medicine; New York Unversity  
**Publication:** Nature medicine letters  
**Year:** 2019  
**Citation:** 125

##### Main idea
This paper proposed a 3D deep learning algorithm which utilizes patients' current and prior CT volumns to localize the cancerous nodules and predict the risk of lung cancer.

##### Problems
Lung cancer screening has low sensitivity and specificity performances and has the inter-grader variability problem which will lead to high clinical and financial costs.

##### Methods
- Framework
<img src="./natmed19letter/Overall modeling framework.PNG" width="80%">  
- Networks
<img src="./natmed19letter/Networks.PNG" width="80%">  


1. Lung segmentation Mask-RCNN trained on LUNA dataset(TensorFlow Object Detection API), only used to find the lung center for alignment in full volume model.
2. Cancer ROI detection using 3D RetinaNet(pretrained first on LIDC then NLST). 
3. 3D inflated Inception V1 to predict cancer (pretrained on ImageNet, with [Focal Loss](https://arxiv.org/pdf/1708.02002.pdf))  
4. Cancer risk prediction model combined both ROIs model(nodule-level local information) and full model(global context ).

##### Results
- When prior CT imaging was not available, proposed model outperformed radiologists with absolute reductions of 11% in FP and 5% in FN. When it is available, model performance was on-par with the radiologists.
- Experiments on NLST dataset (86 cancer out of 6,716 cases) achieved AUC-ROC 94.4%.

##### Take away
1. There are estimated **160,000 deaths** in USA in 2018 because of Lung cancer. It is the most common cause of cancer death.
2. Lung cancer screening using low-dose computed tomography can reduce mortality by **20-43%**.
3. Radiologist workflow:
    - full assessment of low-dose CT volume 
    - focuse on regions of concern
    - comparison to prior imaging when available
    - calibration against biopsy-confirmed outcomes
4. National Lung Cancer Screening Trial (NLST-Kaggle17): 42,290 CT cases from 14,851 patients. 578 patients developed cancer within 1 year. Slice spacing varying from 1.25 to 5 mm. [NLST](https://cdas.cancer.gov/learn/nlst/images/)
5. The Lung Image Database Consortium and Image Database Resource Initialtive(LIDC-IDRI) [LIDC-IDRI](https://wiki.cancerimagingarchive.net/display/Public/LIDC-IDRI)
6. LUng Nodule Analysis (use LIDC/IDRI dataset) [LUNA](https://luna16.grand-challenge.org/data/)
7. [Lung-RADS](https://www.acr.org/-/media/ACR/Files/RADS/Lung-RADS/LungRADSAssessmentCategoriesv1-1.pdf?la=en): Lung CT Screening Reporting & Data System.

#### [Incidental Pulmonary Nodules Detected on CT Images](IncidentalPulm.pdf)

**Authors:** Irsk J. Anderson, MD; Andrew M. Davis, MD, MPH   
**Institutes:** University of Chicago  
**Publication:** JAMA Clinical Guidelines Synopsis  
**Year:** 2018  
**Citation:** 8

##### Main idea
Provide guidlines for management of incidental pulmonary nodules detected on CT images.

##### Problems
An incidental lung nodule on CT scan can create uncertainty for clinicians and anxiety for patients.

##### Methods
Empirical. All the risk factors for malignancy can be found in [guidelines](incidentalPulm/Guidelines.pdf):
1. Nodule Size and Morphology
2. Nodule Location
3. Nodule Multiplicity
4. Nodule Growth Rate
5. Emphysema and Fibrosis
6. Age, Sex, Race, and Family History
7. Tobacco and Other Inhaled Carcinogens
8. Risk Estimation and Risk Models
9. Invasive Diagnostic and Therapeutic Procedures

##### Take away
1. Lung nodule was detected in up to **51%** of study participants in a systematic review of CT screening lung cancer trials.

3. Lung cancer is rare in children and adults younger than 35 years,

#### [Multi-crop Convolutional Neural Networks for lung nodule malignancy suspiciousness classification](multicropCNN.pdf)

**Authors:** W Shen et al.   
**Institutes:** Chinese Academy of Sciences; Beijing Key Laboratory of Molecular Imaging; Stanford University; Beijing Jiaotong University  
**Publication:** Pattern Recognition    
**Year:** 2017  
**Citation:** 185

##### Main idea
Modeling new nodule patches and build and end-to-end multi-crop CNN to classify lung nodule malignancy suspiciousness.

##### Problems
1. From the reports from the experienced thoracic radiologists, each nodule is rated for malignancy suspiciousness from 1 to 5. Nodules with average score lower than 3 and greater than 3 are labelled as **low malignancy-suspicious nodules(LMNs)** and **high malignancy-suspicious nodules(HMNs)** respectively. The task is to correctly classify the malignancy-suspicious level.
2. Classify subtlety and margin of the nodule:
    - Margin: description of how well-defined the margin is
    - Subtlety: difficulty of detection
3. Nodule diameter estimation

##### Methods
- Framework

##### Results
1. Achieved state-of-art nodule suspiciousness classification performance (**Accuracy:87.14%**, AUC 0.93, sensitivity 0.77, specificity 0.93).
2. Able to characterize nodule semantic attributes(subtlety and margin) and nodule diameter.

##### Take away
1. 5-year sruvival rate is around 18% for Lung cancer.
2. Lung cancer is the leading death of cancer (**27%** of all cancer deaths)
